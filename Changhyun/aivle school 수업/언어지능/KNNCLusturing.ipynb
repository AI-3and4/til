```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentence = ('The weather is extremely pleasant and sunny today.', 'The weather is terribly unpleasant and stormy today.')
vector = TfidfVectorizer(max_features=100)
tfidf_vector = vector.fit_transform(sentence)

# print(tfidf_vector[0])
# print(tfidf_vector[1])

print(cosine_similarity(tfidf_vector[0], tfidf_vector[1]))
```

```python
import numpy as np

# 그룹 정의
group_A = np.random.uniform(100, 200, (15, 2))
group_B = np.random.uniform(0, 50, (15, 2))
group_C = np.random.uniform(100, 200, (15, 2))

# 거리 계산 함수
def distance(point, group):
    return np.linalg.norm(point - np.mean(group, axis=0))

point = np.array([120, 30])

# 각 그룹과의 거리 계산
dist_A = distance(point, group_A)
dist_B = distance(point, group_B)
dist_C = distance(point, group_C)

# 가장 가까운 그룹을 출력
distances = [dist_A, dist_B, dist_C]
index = np.argmin(distances)

if index == 0:
    print("Group A")
elif index == 1:
    print("Group B")
else:
    print("Group C")
```

```python
import numpy as np

# 그룹 A, B, C 데이터 초기화
group_A = np.random.uniform(100, 200, (15, 2))
group_B = np.random.uniform(0, 50, (15, 2))
group_C = np.random.uniform(100, 200, (15, 2))

# 각 그룹의 중심과 우리가 가지고 있는 현재 point와의 거리를 구하는 함수
def distance(point, group):
    return np.linalg.norm(point - group, axis=1)

k = 5  # 최근접 이웃 개수

# 그룹의 통계치를 나타내는 리스트
stat = [0, 0, 0]  # 제일 앞은 A에 해당하는 개수, 그 다음은 B, 마지막은 C에 해당하는 개수

# 가장 가까운 k개의 점 중 어느 그룹이 가장 많은지를 계산
index = np.argmax(stat)

if index == 0:
    print("Group A")
elif index == 1:
    print("Group B")
else:
    print("Group C")
```
