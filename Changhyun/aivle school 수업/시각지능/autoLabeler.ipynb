```python
from ultralytics import YOLO
model = YOLO("yolov11n-face.pt")
```

```python
import zipfile

# 압축 풀기
output_dir = './my_images2'
os.makedirs(output_dir, exist_ok=True)

zip_path = '/content/drive/MyDrive/mini_project4/normal_pictures.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(output_dir)
    print(f"{zip_path} -> {output_dir} extreact.")
```

```python
import random

test_dir = './my_images2'

image_files = [f for f in os.listdir(test_dir) if f.lower().endswith(('.jpg', '.jpeg', '.gif', '.bmp'))]

# 무작위로 파일 선택
selected_files = random.sample(image_files, min(10, len(image_files)))
```

```python
import cv2
from matplotlib import pyplot as plt
```

```python
output_folder = './labels'
# os.makedirs(output_folder)

for filename in image_files:
    # 이미지 로드
    img = cv2.imread('./my_images2/' + filename)
    img_height, img_width = img.shape[:2]
    results = model('./my_images2/' + filename, verbose=True)

    # 라벨 파일 생성
    label_file = os.path.join(output_folder, os.path.splitext(filename)[0] + ".txt")
    with open(label_file, "w") as f:
        for box in results[0].boxes:
            x, y, w, h = box.xywh[0]

            # 정규화된 YOLO 형식으로 변환
            x_center = x / img_width
            y_center = y / img_height
            width = w / img_width
            height = h / img_height

            # 클래스 ID (얼굴 감지의 경우 0으로 가정)
            class_id = 0
            f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

            # 사각형 그리기
            start_point = (int(x - w / 2), int(y - h / 2))
            end_point = (int(x + w / 2), int(y + h / 2))
            cv2.rectangle(img, start_point, end_point, (0, 255, 0), 2)

            print(f"Detected box for {filename}: x={x}, y={y}, w={w}, h={h}")
        # img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # plt.figure(figsize=(8, 8))
        # plt.imshow(img_rgb)
        # plt.axis("off")
        # plt.show()
```
